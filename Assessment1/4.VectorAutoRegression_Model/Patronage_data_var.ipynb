# **Predictive model : VAR alias Vector Auto Regression**

A VAR model, or Vector Autoregression model, is a powerful tool for forecasting multiple time series that influence each other, in contrast to traditional models that focus on single series. Imagine you're analyzing prices of stocks A and B. A VAR model can capture how changes in A affect B and vice versa, leading to more accurate forecasts for both.

Here's a breakdown of the key aspects of a VAR model:

**Multiple variables:** As mentioned, VAR models handle two or more time series simultaneously. This allows you to understand the interdependencies between them and how they evolve together.

**Autoregressive nature**: Each variable in the model is explained by its past values (called lags) and the past values of all other variables. This reflects the idea that future values are influenced by what happened previously in the system.

**Equation systems:** A VAR model consists of one equation for each variable, where the left side represents the future value and the right side includes lagged values of all variables and a constant term.



# **Strengths of VAR model**

Captures interdependencies between multiple time series.

Relatively easy to understand and implement.

Flexible and adaptable to various data types.

# **Weaknesses of VAR model**

Can be computationally expensive for large datasets.

Requires careful selection of the lag order (number of past values used).

Sensitive to outliers and structural breaks in the data.

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC')
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)
    for key,val in result[4].items():
        out[f'critical value ({key})']=val
    print(out.to_string())
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("ts_input_data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("ts_input_data has a unit root and is non-stationary")
```

# **Checking the stattionarity using the hypothesis testing of all the factors involved**

```python
adf_test(ts_input_data['MyWay'])
print("\n")
adf_test(ts_input_data['Paper Ticket'])
print("\n")
adf_test(ts_input_data['Total'])
```

# **Splitting the data for training & evaluation**


Splitting time series data for training and testing requires a different approach than standard data because of the inherent temporal dependence between observations. Randomly splitting won't work here, as it disrupts the time order and creates unrealistic scenarios for testing. Here are some methods for efficiently splitting time series data:

**1. Simple Fixed Split:**

This is the easiest method, where you define a percentage of the data for training and the remaining for testing. For example, 80% for training and 20% for testing.
While simple, it might not be the most statistically sound, as the chosen split point might influence the model's performance.


**2. Time-based Split:**

This method respects the temporal order of your data. You define a specific date or timestamp as the split point, ensuring the training data is chronologically before the testing data.
This approach is effective for simulating real-world situations where you wouldn't have access to future data for training.


**3. Rolling Window Split:**

This method involves creating multiple training and testing sets by sequentially moving a window of data across the entire time series.
Each window becomes a separate train-test pair, leading to a more robust assessment of the model's performance across different time periods.


# **Splitting the data in 3:1 ratio (75%:25%) for training & testing**

```python
train = ts_input_data[:int(0.75*(len(ts_input_data)))]
valid = ts_input_data[int(0.25*(len(ts_input_data))):]
```

# **Implementing VAR model using sklearn**

```python
from statsmodels.tsa.vector_ar.var_model import VAR
# Fit the model
model = VAR(endog=train, freq=None, missing='none')
model_fit = model.fit()
print(model_fit)
```

```python
# Make prediction on validation
prediction = model_fit.forecast(model_fit.endog, steps=len(valid))
```

# **Determining the Lag Values**

In a VAR (Vector Autoregression) model, a lag value refers to the value of a variable at a previous time period. It's a core concept that captures the temporal dependencies within the model.

Here's a breakdown of how lags work in VAR models:

**1. Understanding Lag Order:**

The number of lags included in a VAR model is called its "lag order" and is denoted by p.
A VAR(p) model means it incorporates lags for the last p time periods.

**2. Incorporating Lags in Equations:**

Each variable in a VAR model has its own equation.
The right-hand side of each equation includes both:
Lagged values of the variable itself (its own past values).
Lagged values of all other variables in the model.


**3. Example:**

Consider a VAR(2) model with two variables, x and y:

Equation for x:
x_t = a1 + b11*x_(t-1) + b12*y_(t-1) + b13*x_(t-2) + b14*y_(t-2) + Îµ_t
Equation for y:
y_t = a2 + b21*x_(t-1) + b22*y_(t-1) + b23*x_(t-2) + b24*y_(t-2) + Î·_t
In this example:

x_t and y_t represent the current values of the variables.
x_(t-1), y_(t-1), x_(t-2), and y_(t-2) are the lag values from the previous two time periods.
The coefficients (b11, b12, etc.) quantify the effect of each lag on the current value.


**4. Importance of Lag Selection:**

Choosing the appropriate lag order is crucial for VAR model accuracy.
Too few lags might miss important relationships, while too many can lead to overfitting and poor out-of-sample performance.
Statistical tests and information criteria are often used to guide lag selection.



Lag values enable VAR models to capture the dynamic relationships between multiple time series variables.
They reflect the idea that future values of a variable are influenced by its own past and the past of other variables in the system.
